```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from sklearn import linear_model
```

```python
#importing dataset
df = pd.read_csv("C:\\Users\\91917\\Desktop\\Datasets\\homeprices.csv")
df
```

```python
%matplotlib inline 
plt.xlabel('Area')
plt.ylabel('Price')
plt.scatter(df.Area, df.Price, color="Green", marker="*")
plt.show()
```

```python
new_df = df.drop('Price',axis='columns')
new_df
```

```python
price = df.Price 
price
```

```python
# create linear reg model 
reg = linear_model.LinearRegression()
reg.fit(new_df,price)
```

# Predict price of home with area = 3300 sa ft

```python
reg.predict([[3300]]) #prdicting price of 3300 sq ft area 
```

```python
reg.coef_             
```
